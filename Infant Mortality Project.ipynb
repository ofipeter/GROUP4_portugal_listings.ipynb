# 7. global-vaccination-coverage

```python
df7
```

```python
df7.describe()
```

```python
df7.isnull().sum()
```

```python
# I drop Code column and other columns with 50% or more null sets
df7 = df7.drop(columns=[
    'Code', 'Hib3 (% of one-year-olds immunized)', 'IPV1 (% of one-year-olds immunized)', 'PCV3 (% of one-year-olds immunized)',
    'RotaC (% of one-year-olds immunized)', 'YFV (% of one-year-olds immunized)'
], axis=1)
df7
```

```python
# Fill all NaN with mean values
df7 = df7.fillna(df7.mean())
df7
```

```python
# Group each of the Vaccines by Entity and Year given
group_df7_BCG = df7.groupby(['Entity', 'Year'])['BCG (% of one-year-olds immunized)'].sum()
print("Total BCG:\n", group_df7_BCG)

group_df7_Hep = df7.groupby(['Entity', 'Year'])['HepB3 (% of one-year-olds immunized)'].sum()
print("Total Hep:\n", group_df7_Hep)

group_df7_MCV = df7.groupby(['Entity', 'Year'])['MCV1 (% of one-year-olds immunized)'].sum()
print("Total MCV:\n", group_df7_MCV)

group_df7_Pol = df7.groupby(['Entity', 'Year'])['Pol3 (% of one-year-olds immunized)'].sum()
print("Total Polio:\n", group_df7_Pol)

group_df7_RCV = df7.groupby(['Entity', 'Year'])['RCV1 (% of one-year-olds immunized)'].sum()
print("Total RCV:\n", group_df7_RCV)

group_df7_DTP = df7.groupby(['Entity', 'Year'])['DTP3 (% of one-year-olds immunized)'].sum()
print("Total DTP:\n", group_df7_DTP)
```

```python
# Highest vacinattion given in Entity and Year it was given
highest_BCG = group_df7_BCG.idxmax()
print("Highest Entity and Year where BCG was Given:\n", highest_BCG)

highest_Hep = group_df7_Hep.idxmax()
print("Highest Entity and Year with HEP Vaccine:\n", highest_Hep)

highest_MCV = group_df7_MCV.idxmax()
print("Entity and Year with Highest MCV Vacination:\n", highest_MCV)

highest_Pol = group_df7_Pol.idxmax()
print("Entity and Year with Highest Polio Vaccination:\n", highest_Pol)

highest_RCV = group_df7_RCV.idxmax()
print("Entity and Year with Highest RCV Vaccination:\n", highest_RCV)

highest_DTP = group_df7_DTP.idxmax()
print("Entity and Year with Highest DTP Vaccination:\n", highest_DTP)
```
